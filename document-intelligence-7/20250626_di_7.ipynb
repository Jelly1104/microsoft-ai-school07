```python
import requests
import gradio as gr


def requests_document_intelligence(imaage_path):
    endpoints="https://7ai053-document-intelligence.cognitiveservices.azure.com/documentintelligence/documentModels/prebuilt-read:analyze?_overload=analyzeDocument&api-version=2024-11-30"
    headers={"Ocp-Apim-Subscription-Key":"BHifxxlNBZpytKj8d6PAC2iYbo7z8DNPGkDGpfzmzhhZtdDrjAxeJQQJ99BFACYeBjFXJ3w3AAALACOGc5EC","Content-Type":"image/png"}
 
    with open(imaage_path, "rb") as image_file: #binary 형태로 Image를 읽음
        image_data=image_file.read()

    response_post=requests.post(endpoints,headers=headers,data=image_data)

    if response_post.status_code != 202:
        print(f"Error:{response_post.status_code}")
        return None
    
    url=response_post.headers['operation-location']
    print(url)


    while True:

        result_response=requests.get(url,headers=headers)

        if result_response.status_code != 200:
            print(f"Error:{result_response.status_code}")
            return None
        
        result_response_json=result_response.json()
        current_status=result_response_json.get("status")

        
        if current_status == "running":
            print("current Status : {}".format(current_status))
            continue
        else:
            break

        # 요청을 보낸 후 받은 응답의 상태(status) 값이 succeeded인지 확인한다.
        # running이라면, 다시 GET 요청을 보낸다.

    if current_status == "succeeded":
        # print(result_response_json)
        return result_response_json
    else:
        return None

    # succeeded가 아니라면, 다시 GET 요청을 보낸다.
    # succeeded라면, 응답의 결과를 출력한다.


def randomcolor():
    import random
    return (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))


def get_font():
    from PIL import ImageFont
    import platform
    
    font_size=20

    try:
        if platform.system() == "windows":
            return ImageFont.truetype("malgun.ttf",font_size)
        elif platform.system() == "Darwin": # MAC OS
            return ImageFont.truetype("AppleGothic",font_size)
        else:   # Linux
            return ImageFont.load_default(size=font_size)
    except IOError:
        return ImageFont.load_default(size=font_size)


def draw_polygon(image_path,response_data):
    from PIL import Image, ImageDraw
    
    image=Image.open(image_path)
    draw = ImageDraw.Draw(image)

    block_list=response_data['analyzeResult']['paragraphs'] # response_data에서 polygon 정보를 추출해서,
    for block in block_list:
        line_color = randomcolor()
        font=get_font()
        polygon = block['boundingRegions'][0]['polygon']
        # content 정보 추출
        content=block['content']
        # (x, y)튜플 리스트로 변환
        polygon_points = [(polygon[i], polygon[i+1]) for i in range(0, len(polygon), 2)]

        draw.polygon(polygon_points, outline=line_color,width=2)
        draw.text((polygon[0],polygon[1]-20),content,fill=line_color,font=font)

        
    return image


with gr.Blocks() as demo:

    def change_image(image_path):
        response_data = requests_document_intelligence(image_path)        # 
        image=draw_polygon(image_path, response_data)
        return image
    
    with gr.Row():
        input_image=gr.Image(label="이미지선택",type="filepath", width=500)
        output_image=gr.Image(label="결과 이미지",type="pil",interactive=False, width=500)
        # output_image=gr.Image(label="결과 이미지",type="pil",interactive=False)

    input_image.change(change_image,inputs=[input_image],outputs=[output_image])
demo.launch(share=True)


# change_image('/Users/eunahjeong/Desktop/doit/document-intelligence-7/20250626_dl_7_image.jpg')
```
